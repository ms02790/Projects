```python
# initializing rows and columns
Row_Number = 3
Column_Number = 3

#Defining state of the game
class Gamestatus:
    # vacant board and two players 
    def __init__(selfinstructor, player1, player2):
        #vacating the board
        selfinstructor.tictack = np.zeros((Row_Number, Column_Number))
        #the two players
        selfinstructor.player1 = player1
        selfinstructor.player2 = player2
        selfinstructor.isEnding = False
        selfinstructor.Hashboard = None
        # player1 plays first
        selfinstructor.symbolply = 1

    # hashes the current state of the board
    def Hashvalue(selfinstructor):
        selfinstructor.Hashboard = str(selfinstructor.tictack.reshape(Column_Number * Row_Number))
        return selfinstructor.Hashboard

    #judging the winner
    def win_condition(selfinstructor):
        
        # column Number
        for i in range(Column_Number): 
            if sum(selfinstructor.tictack[:, i]) == 3:
                selfinstructor.isEnding = True
                # for player 1
                return 1
            if sum(selfinstructor.tictack[:, i]) == -3:
                selfinstructor.isEnding = True
                # for player 2
                return -1
        # row Number
        for i in range(Row_Number):
            if sum(selfinstructor.tictack[i, :]) == 3:
                selfinstructor.isEnding = True
                #for player 1
                return 1
            if sum(selfinstructor.tictack[i, :]) == -3:
                selfinstructor.isEnding = True
                #for player 2
                return -1
        
        # draw
        if len(selfinstructor.Positionsavailable()) == 0:
            selfinstructor.isEnding = True
            # game outcome 0
            return 0
        
        selfinstructor.isEnding = False
        #None if game not end
        return None
    
        # diagonal
        diagonal_summation1 = sum([selfinstructor.tictack[i, i] for i in range(Column_Number)])
        diagonal_summation2 = sum([selfinstructor.tictack[i, Column_Number - i - 1] for i in range(Column_Number)])
        diagonal_summation = max(abs(diagonal_summation1), abs(diagonal_summation2))
        if diagonal_summation == 3:
            selfinstructor.isEnding = True
            if diagonal_summation1 == 3 or diagonal_summation2 == 3:
                #player 1
                return 1
            else:
                #player 2
                return -1

    # estimation after each game.
    def Gamestate_updation(selfinstructor, position):
        #position on the board
        selfinstructor.tictack[position] = selfinstructor.symbolply
        
        #switching between players
        selfinstructor.symbolply = -1 if selfinstructor.symbolply == 1 else 1
    
    
    def Positionsavailable(selfinstructor):
        # Position of the variable
        posit = []
        for i in range(Row_Number):
            for j in range(Column_Number):
                if selfinstructor.tictack[i, j] == 0:
                    posit.append((i, j)) 
        return posit

    
    # When game finishes
    def finalresult(selfinstructor):
        # game outcome
        result = selfinstructor.win_condition()
        if result == 1:
            # player 1 wins
            selfinstructor.player1.OfferReward(1)
            selfinstructor.player2.OfferReward(0)
        elif result == -1:
            # player 2 wins
            selfinstructor.player1.OfferReward(0)
            selfinstructor.player2.OfferReward(1)
        else:
            # ends with a draw
            selfinstructor.player1.OfferReward(0.1)
            selfinstructor.player2.OfferReward(0.5)

    # board resetting_board
    def resetting_board(selfinstructor):
        # resetting all values
        selfinstructor.tictack = np.zeros((Row_Number, Column_Number))
        selfinstructor.Hashboard = None
        selfinstructor.isEnding = False
        selfinstructor.symbolply = 1

    def perf(selfinstructor, iterations=100):
        for i in range(iterations):
            if i % 1000 == 0:
                print("iterations {}".format(i))
            while not selfinstructor.isEnding:
                
                positions = selfinstructor.Positionsavailable()
                player1_action = selfinstructor.player1.Actionchoice(positions, selfinstructor.tictack, selfinstructor.symbolply)
                # updation on gamestate
                selfinstructor.Gamestate_updation(player1_action)
                board_hash = selfinstructor.Hashvalue()
                selfinstructor.player1.addGameState(board_hash)
                # if game ends check status

                win = selfinstructor.win_condition()
                if win is not None:
                    # Player 1
                    selfinstructor.finalresult()
                    selfinstructor.player1.resetting_board()
                    selfinstructor.player2.resetting_board()
                    selfinstructor.resetting_board()
                    break

                else:
                    # Player 2
                    positions = selfinstructor.Positionsavailable()
                    player2_action = selfinstructor.player2.Actionchoice(positions, selfinstructor.tictack, selfinstructor.symbolply)
                    selfinstructor.Gamestate_updation(player2_action)
                    board_hash = selfinstructor.Hashvalue()
                    selfinstructor.player2.addGameState(board_hash)

                    win = selfinstructor.win_condition()
                    if win is not None:
                        
                        selfinstructor.finalresult()
                        selfinstructor.player1.resetting_board()
                        selfinstructor.player2.resetting_board()
                        selfinstructor.resetting_board()
                        break

    
    def perf2(selfinstructor): 
        while not selfinstructor.isEnding:
            # Player 1
            positions = selfinstructor.Positionsavailable()
            player1_action = selfinstructor.player1.Actionchoice(positions, selfinstructor.tictack, selfinstructor.symbolply)
            # Updation of state
            selfinstructor.Gamestate_updation(player1_action)
            selfinstructor.setshowcase()
            # Checking the status of the board
            win = selfinstructor.win_condition()
            if win is not None:
                if win == 1:
                    print(selfinstructor.player1.name, "wins!")
                else:
                    print("tie!")
                selfinstructor.resetting_board()
                break

            else:
                # Player 2
                positions = selfinstructor.Positionsavailable()
                player2_action = selfinstructor.player2.Actionchoice(positions)

                selfinstructor.Gamestate_updation(player2_action)
                selfinstructor.setshowcase()
                win = selfinstructor.win_condition()
                if win is not None:
                    if win == -1:
                        print(selfinstructor.player2.name, "wins!")
                    else:
                        print("tie!")
                    selfinstructor.resetting_board()
                    break

    def setshowcase(selfinstructor): 
        for i in range(0, Row_Number):
            print('----------------')
            op = '| '
            for j in range(0, Column_Number):
                if selfinstructor.tictack[i, j] == 1:
                    # player1:-x 
                    token = 'x'
                if selfinstructor.tictack[i, j] == -1:
                    # player2:-o
                    token = 'o'
                if selfinstructor.tictack[i, j] == 0:
                    token = ' '
                op += token + ' | '
            print(op)
        print('---------------')


```

```python
class Competitor:
    def __init__(selfinstructor, name, exrate=0.3):
        selfinstructor.name = name
        selfinstructor.Gamestatuss = []  
        selfinstructor.lr = 0.2
        selfinstructor.exrate = exrate
        selfinstructor.decay_gamma = 0.9
        selfinstructor.Gamestatuss_value = {}  

    def Hashvalue(selfinstructor, board):
        Hashboard = str(board.reshape(Column_Number * Row_Number))
        return Hashboard

    def Actionchoice(selfinstructor, posit, currentset, symbol):
        if np.random.uniform(0, 1) <= selfinstructor.exrate:
            # take random action
            indx = np.random.choice(len(posit))
            action = posit[indx]
        else:
            maxvalue = -999
            for k in posit:
                nextset = currentset.copy()
                nextset[k] = symbol
                next_Hashboard = selfinstructor.Hashvalue(nextset)
                value = 0 if selfinstructor.Gamestatuss_value.get(next_Hashboard) is None else selfinstructor.Gamestatuss_value.get(next_Hashboard)
                # print("value", value)
                if value >= maxvalue:
                    maxvalue = value
                    action = k
        # print("{} takes action {}".format(selfinstructor.name, action))
        return action
    
    # backpropogate
    def OfferReward(selfinstructor, rewards):
        for st in reversed(selfinstructor.Gamestatuss):
            if selfinstructor.Gamestatuss_value.get(st) is None:
                selfinstructor.Gamestatuss_value[st] = 0
            selfinstructor.Gamestatuss_value[st] += selfinstructor.lr * (selfinstructor.decay_gamma * rewards - selfinstructor.Gamestatuss_value[st])
            rewards = selfinstructor.Gamestatuss_value[st]


    # appending a hash GameState
    def addGameState(selfinstructor, Gamestatus):
        selfinstructor.Gamestatuss.append(Gamestatus)

   
    def resetting_board(selfinstructor):
        selfinstructor.Gamestatuss = []

    def Policysaving(selfinstructor):
        filewrite = open('policy_' + str(selfinstructor.name), 'wb')
        pickle.dump(selfinstructor.Gamestatuss_value, filewrite)
        filewrite.close()

    def Policyloading(selfinstructor, file):
        fileread = open(file, 'rb')
        selfinstructor.Gamestatuss_value = pickle.load(fileread)
        fileread.close()


```
